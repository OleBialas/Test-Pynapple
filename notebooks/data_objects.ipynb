```python
import numpy as np
import pandas as pd
import pynapple as nap
```

## One dimensional time-series data

Generate numpy arrays for data `d` and time stamps `t`

```python
rng = np.random.default_rng(7)
d = rng.random(100)
t = np.linspace(0, 1, 100)
```

Create time series data `Tsd` object.

```python
tsd = nap.Tsd(t, d)
tsd
```

The object automatically calculates the sampling `rate` from the time stamps.

```python
tsd.rate
```

The data can be exported to a numpy array

```python
d_tsd = tsd.as_array()
np.testing.assert_equal(d_tsd, d)
```

It can also be exported to a pandas series. Unlike the numpy array, this includes the timestamps.

```python
series = tsd.as_series()
```

The Tsd object can also be intialized from a series. In this case, we do not need to provide `d`. This assumes that the pandas index represents time.

```python
s = pd.Series(data=[0.2, 0.4, 0.1], index=[0.0, 0.5, 1.0])
nap.Tsd(t=s)
```

When timestamps are not sorted, pynapple sorts them and throws a warning

```python
s = pd.Series(data=[0.2, 0.4, 0.1], index=[0.5, -0.1, 1])
nap.Tsd(t=s)
```

## Multi-dimensional time series data

2-d data (e.g. mutiple channels) can be represented in a `TsdFrame` object with optional names for `columns`.

```python
rng = np.random.default_rng(7)
d = rng.random((100, 3))
t = np.linspace(0, 1, 100)

tsd = nap.TsdFrame(t, d, columns=["ch1", "ch2", "ch3"])
tsd
```

The Tsd objects can be sliced like arrays

```python
tsd[:,0]
```

Numpy functions also work the same

```python
np.mean(tsd, axis=0)
```

We can also select a column like in pandas using `.loc`

```python
tsd.loc["ch2"]
```

Higher dimensional data (e.g. movies) can be represented with the `TsdTensor` object.

```python
rng = np.random.default_rng(7)
d = rng.random((100, 3, 4))
t = np.linspace(0, 1, 100)
```

```python
tsd = nap.TsdTensor(t, d)
tsd
```

## Restricting continuous data to epochs with intervals

We can store the start and stop of intervals (e.g. stimuli) in an `IntervalSet`.

```python
ep = nap.IntervalSet(start=[0, 0.25, 0.5], end=[0.1, 0.35, 0.6])
ep
```

When the end prededs the start, the interval is dropped with a warning.

```python
ep = nap.IntervalSet(start=[0, 0.25, 0.5], end=[0.1, 0.35, 0.4])
ep
```

By using the `.restrict` method, we can select only the samples that fall into the intervals

```python
rng = np.random.default_rng(7)
d = rng.random((100, 3))
t = np.linspace(0, 1, 100)

tsd = nap.TsdFrame(t, d, columns=["ch1", "ch2", "ch3"])
tsd.restrict(ep)
```

The restriction can also be applied when creating the Tsd objects using the `time_support` argument.

```python
tsd = nap.TsdFrame(t, d, columns=["ch1", "ch2", "ch3"], time_support=ep)
tsd
```

The intervals are stored in the `.time_support` attribute.

```python
tsd.time_support
```

## Representing spike times with time stamps

Discrete time stamps can be stored in the `Ts` class.

```python
rng = np.random.default_rng(7)
t = np.sort(rng.random(100))
ts = nap.Ts(t)
ts
```

The Ts can also be resticted using intervals

```python
ts.restrict(ep)
```

Multiple spike trains can be represented as a `TsGroup` which is created from a dataframe of 

```python
rng = np.random.default_rng(7)
data = {n: nap.Ts(np.sort(rng.random(n*5))) for n in range(10)}
ts = nap.TsGroup(data)
ts
```
